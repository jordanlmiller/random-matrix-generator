```python
n = 5

r = np.random.random((n,n))
ortho = util.column_gram_schmidt(r)
rows, cols = ortho.shape

for i in range(rows):
    for j in range(rows):
        print(i, j, np.around(inner(ortho[i,:], ortho[j,:])))

for i in range(cols):
    for j in range(cols):
        print(i, j, np.around(inner(ortho[:,i], ortho[:,j])))
```

```python
import matplotlib.pyplot as plt
gen = Complex(3)

angle = np.ndarray(1000)
for i in range(1000):
    u = gen.special_unitary()
    d = np.linalg.det(u)
    a,b = d.real, d.imag
    angle[i] = np.arctan2(a,b)
    print(np.arctan2(a,b))

plt.hist(angle)
```
